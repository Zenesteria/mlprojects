# Module Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Cost Function
# $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=100)
x_1
```

```python
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1)) #start with x values then y values
```

# Cost Function Derivative
Represents the slope of the graph

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[10, 5])

plt.subplot(1, 2, 1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1)) #start with x values then y values

plt.subplot(1, 2, 2)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1)) #start with x values then y values
```

# GRADIENT DESCENT ALGORITHM

```python
def gradient_descent(newVal, stepMultiplier, precision):
    prev_val = 0
    step_size = abs(newVal - prev_val) 
    x_values = [newVal]
    y_values = [df(newVal)]
    while step_size > precision:
        prev_val = newVal
        gradient = df(prev_val)
        newVal = prev_val - stepMultiplier*gradient
        x_values.append(newVal)
        y_values.append(df(newVal))
        step_size = abs(newVal - prev_val) 

    return [x_values, y_values]

coordinates = gradient_descent(3, 0.1,0.0001)
    
```

```python
plt.figure(figsize=[10, 5])

plt.subplot(1, 3, 1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1)) #start with x values then y values
coordinatesFormatted = np.array(coordinates[0])
plt.scatter(coordinates[0], f(coordinatesFormatted), color='red')

plt.subplot(1, 3, 2)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.scatter(coordinates[0], coordinates[1], color='red')
plt.plot(x_1, df(x_1)) #start with x values then y values

plt.subplot(1, 3, 3)
plt.xlim([-0.2,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1)) #start with x values then y values
plt.scatter(coordinates[0], coordinates[1], color='red')
```
