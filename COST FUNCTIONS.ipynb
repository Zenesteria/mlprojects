# Module Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Cost Function

# $$  f(\theta) = (\theta - 2)^2  $$


```python
def f(theta):
    return (theta - 2)**2
```

## Data Input

```python
theta_one = np.linspace(start=-5, stop=5, num=1000)
```

```python
plt.figure(figsize=[12,5])
plt.title('Cost Function')
plt.ylabel('f(theta)', fontsize=18)
plt.xlabel('theta', fontsize=18)

plt.grid()
plt.plot(theta_one, f(theta_one), color='turquoise')
```

## Slope of the Cost Function

```python
def df(theta):
    return 2*(theta - 2)
```

```python
plt.figure(figsize=[12,5])



plt.subplot(1, 2, 1)
plt.grid()
plt.title('Cost Function', fontsize=20)
plt.xlabel('theta', fontsize=18)
plt.ylabel('f(theta)', fontsize=18)
plt.plot(theta_one, f(theta_one), color='turquoise')

plt.subplot(1, 2, 2)
plt.grid()
plt.title('Slope of the Cost Function', fontsize=20)
plt.xlabel('theta', fontsize=18)
plt.ylabel('df(theta)', fontsize=18)
plt.plot(theta_one, df(theta_one), color='m')
```

## Finding the Minimum Cost Value Using Gradient Descent

```python
def gradient_descent(derivative_func, start_value, iterations, learning_rate=0.1, precision=0.001):
    # Create Lists to store the point values
    new_values = [start_value]
    new_derivative_values = [derivative_func(start_value)]

    # The function loops through based on the number of iterations
    for n in range(iterations):
        prev_value = start_value
        start_value = prev_value - derivative_func(prev_value)*learning_rate
        
        new_values.append(start_value)
        new_derivative_values.append(derivative_func(start_value))
        
        step_size = abs(start_value - prev_value)

        # Break the loop if the difference between the previous and current value is lower than the precision set
        if(step_size < precision):
            break

    return start_value, new_values, new_derivative_values
    
```
