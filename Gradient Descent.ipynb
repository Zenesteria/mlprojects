# Cost Function
# $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

# Cost Function Derivative
# $ 2x + 1 $

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[10, 5])

plt.subplot(1, 2, 1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1)) #start with x values then y values

plt.subplot(1, 2, 2)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1)) #start with x values then y values
```

# GRADIENT DESCENT ALGORITHM

```python
def gradient_descent(startVal,derivative, stepMultiplier=0.1, precision=0.000001):
    prev_val = 0
    step_size = abs(startVal - prev_val) 
    x_values = [startVal]
    y_values = [derivative(startVal)]
    
    # while the step size which is the 
    # difference between the old and new val
    # is greater than the precision it should keep 
    # reducing the value of the starting value
    while step_size > precision:
        prev_val = startVal
        gradient = derivative(prev_val)
        startVal = prev_val - stepMultiplier*gradient
        x_values.append(startVal)
        y_values.append(derivative(startVal))
        step_size = abs(startVal - prev_val) 

    return x_values, y_values


    
```

```python
x_vals, grad_vals = gradient_descent(startVal=3, derivative=df)

plt.figure(figsize=[13, 5])

plt.subplot(1, 3, 1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1)) #start with x values then y values
coordinatesFormatted = np.array(x_vals)
plt.scatter(x_vals, f(coordinatesFormatted), color='red')

plt.subplot(1, 3, 2)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.scatter(x_vals, grad_vals, color='red')
plt.plot(x_1, df(x_1)) #start with x values then y values

plt.subplot(1, 3, 3)
plt.xlim([-0.2,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1)) #start with x values then y values
plt.scatter(x_vals, grad_vals, color='red')
```

# Multiple Minima vs Initial Guess and Advanced Functions
# $ x^4 - 4x^2 + 5 $

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

# Cost Function Gradient
# $4x^3 - 8x$

```python
def dg(x):
    return 4*x**3 - 8*x
```

### Data Input

```python
x_2 = np.linspace(start=-3, stop=3, num=1000)
```

```python
plt.figure(figsize=[12, 5])

# Gradient Descent


plt.subplot(1, 2, 1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2))

plt.subplot(1,2,2)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, dg(x_2))
```

## Data Visualization (Gradient Descent)
# $g(x) = x^4 - 4x^2 + 5, dg(x) = 4x^3 - 8x$

```python
plt.figure(figsize=[10, 5])

# Gradient Descent
x_2_vals, grad_2_vals = gradient_descent(startVal=-0.01, derivative=dg, stepMultiplier=0.02)

plt.subplot(1, 2, 1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2))
plt.scatter(x_2_vals, g(np.array(x_2_vals)), color='red')

plt.subplot(1,2,2)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, dg(x_2))
plt.scatter(x_2_vals, grad_2_vals, color='red')
```

# Divergence, Overflow, and Python Tuples
# $h(x) = x^5 - 2x^4 + 2$

## Cost Function

```python
def h(x):
    return x**5 - 2*x**4 + 2
```

## Derivative Cost Function
## $dh(x) = 5x^4 - 8x^3$

```python
def dh(x):
    return 5*x**4- 8*x**3
```

### Data Input

```python
x_3 = np.linspace(start=-2.5, stop=2.5, num=1000)
```

```python
plt.figure(figsize=[13, 5])

# Gradient Descent
x_3_vals, grad_3_vals = gradient_descent(startVal=0.2, derivative=dh, stepMultiplier=0.02)

plt.subplot(1, 2, 1)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.plot(x_3, h(x_3))
plt.scatter(x_3_vals, h(np.array(x_3_vals)), color='red')

plt.subplot(1,2,2)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.xlim(-1, 2)
plt.ylim(-4, 5)
plt.plot(x_3, dh(x_3))
plt.scatter(x_3_vals, grad_3_vals, color='red')
```
